```python
train_data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test_data = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python

for i in range(2):
    image_data = train_data.iloc[i, 1:].values  
    image_array = image_data.reshape(28, 28)

    plt.imshow(image_array, cmap='gray')
    plt.title(f"Label: {train_data.iloc[i, 0]}")
    plt.axis('off')  # Hide the axis
    plt.show()
```

```python
X = train_data.drop(columns=['label'])
y = train_data['label']
y

```

```python
y.value_counts()
```

```python
svm_model = SVC(kernel='rbf', gamma='scale', C=1.0)
```

```python
svm_model.fit(X, y)
```

```python

preds = svm_model.predict(test_data)

preds
```

```python
ids = [i for i in range(1, len(test_data) + 1)]
```

```python
try:
    results = pd.DataFrame({'ImageId': ids, 'Label': preds})
    results.to_csv('submission.csv', index=False)
    print("Succesfully saved!")
except Exception as e:
    print(e)
    print("Something went wrong")
```
